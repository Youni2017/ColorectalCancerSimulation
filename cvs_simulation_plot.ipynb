```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取CSV文件
df = pd.read_csv('output.csv', header=None)

# 获取年龄和癌变状态的概率
years = df[0]
state_331 = df[37]  # (3,3,1)状态
state_341 = df[39]  # (3,4,1)状态 
state_431 = df[47]  # (4,3,1)状态
state_441 = df[49]  # (4,4,1)状态

# 计算总的癌变概率
cancer_prob = state_331 + state_341 + state_431 + state_441

# 创建图表
plt.figure(figsize=(10, 8))

# 使用半对数坐标绘制数据点
plt.semilogy(years, cancer_prob, '+', color='black', label='τ-leaping simulations', markersize=4)

# 设置图表属性
plt.grid(True, which='both', linestyle='-', alpha=0.2)
plt.xlabel('Age (years)')
plt.ylabel('Probability')

# 设置坐标轴范围
plt.xlim(20, 80)
plt.ylim(1e-8, 0.01)

# 添加图例
plt.legend()

# 添加网格
plt.grid(True, which='major', linestyle='-', alpha=0.5)
plt.grid(True, which='minor', linestyle=':', alpha=0.2)

plt.show()

# 打印一些统计信息
print(f"数据点数量: {len(years)}")
print(f"20岁时的癌症概率: {cancer_prob[20]:.2e}")
print(f"50岁时的癌症概率: {cancer_prob[50]:.2e}")
print(f"80岁时的癌症概率: {cancer_prob[80]:.2e}")
```
